```python
# Import required libraries
import pandas as pd
import numpy as np
from google.colab import files
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.vector_ar.vecm import coint_johansen
import matplotlib.pyplot as plt

# List of stocks
stocks = ['AMD', 'INTC', 'AMZN', 'NDAQ', 'TSLA']
data = {}

# Upload Excel files for each stock individually
for stock in stocks:
    print(f"Please upload the Excel file for {stock} (7 days of closing prices)")
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    df = pd.read_excel(filename)

    print(f"\nAvailable columns in {stock} DataFrame:", list(df.columns))
    close_column = input(f"Please enter the column name containing closing prices for {stock}: ")

    try:
        data[stock] = df[close_column]
    except KeyError:
        print(f"Error: Column '{close_column}' not found in {stock} DataFrame")
        raise

# Convert to DataFrame
data_df = pd.DataFrame(data)

# Function for unit root tests with interpretation
def unit_root_tests(series, name):
    print(f"\nUnit Root Tests for {name}:")
    adf_result = adfuller(series)
    print("ADF Test:")
    print(f'ADF Statistic: {adf_result[0]:.4f}')
    print(f'p-value: {adf_result[1]:.4f}')
    print("Interpretation:")
    if adf_result[1] < 0.05:
        print(f"  - p-value < 0.05: Reject null - {name} is stationary")
    else:
        print(f"  - p-value >= 0.05: Fail to reject null - {name} may be non-stationary")

    kpss_result = kpss(series, regression='c')
    print("\nKPSS Test:")
    print(f'KPSS Statistic: {kpss_result[0]:.4f}')
    print(f'p-value: {kpss_result[1]:.4f}')
    print("Interpretation:")
    if kpss_result[1] < 0.05:
        print(f"  - p-value < 0.05: Reject null - {name} is non-stationary")
    else:
        print(f"  - p-value >= 0.05: Fail to reject null - {name} may be stationary")

# Perform unit root tests
for stock in stocks:
    unit_root_tests(data[stock], stock)

# Difference the series
diff_data = data_df.diff().dropna()

# Function to find best AR model (p, 0, 0)
def find_best_ar(series, name, max_p=2):
    best_aic = float('inf')
    best_order = None
    for p in range(max_p + 1):
        try:
            model = ARIMA(series, order=(p, 0, 0))
            results = model.fit()
            if results.aic < best_aic:
                best_aic = results.aic
                best_order = (p, 0, 0)
        except:
            continue
    print(f"\nBest AR model for {name}:")
    print(f"Order: {best_order}")
    print(f"AIC: {best_aic:.2f}")
    return ARIMA(series, order=best_order).fit()

# Fit AR models
ar_models = {stock: find_best_ar(data[stock], stock) for stock in stocks}

# Function to find best ARMA model (d=0) for differenced series
def find_best_arma(series, name, max_p=2, max_q=2):
    best_aic = float('inf')
    best_order = None
    for p in range(max_p + 1):
        for q in range(max_q + 1):
            try:
                model = ARIMA(series, order=(p, 0, q))
                results = model.fit()
                if results.aic < best_aic:
                    best_aic = results.aic
                    best_order = (p, 0, q)
            except:
                continue
    print(f"\nBest ARMA model for {name} (differenced):")
    print(f"Order: {best_order}")
    print(f"AIC: {best_aic:.2f}")
    return ARIMA(series, order=best_order).fit()

# Fit ARMA models
arma_models = {stock: find_best_arma(diff_data[stock], stock) for stock in stocks}

# Function to find best ARIMA model for original series
def find_best_arima(series, name, max_p=2, max_d=1, max_q=2):
    best_aic = float('inf')
    best_order = None
    for p in range(max_p + 1):
        for d in range(max_d + 1):
            for q in range(max_q + 1):
                try:
                    model = ARIMA(series, order=(p, d, q))
                    results = model.fit()
                    if results.aic < best_aic:
                        best_aic = results.aic
                        best_order = (p, d, q)
                except:
                    continue
    print(f"\nBest ARIMA model for {name}:")
    print(f"Order: {best_order}")
    print(f"AIC: {best_aic:.2f}")
    return best_order

# Find and fit ARIMA models
arima_orders = {stock: find_best_arima(data[stock], stock) for stock in stocks}
arima_models = {stock: ARIMA(data[stock], order=arima_orders[stock]).fit() for stock in stocks}

# Cointegration test for pairs (e.g., AMD-INTC)
def cointegration_test_pair(df, stock1, stock2):
    pair_df = df[[stock1, stock2]].dropna()
    try:
        result = coint_johansen(pair_df, det_order=0, k_ar_diff=1)
        print(f"\nJohansen Cointegration Test for {stock1} and {stock2}:")
        print(f"Trace statistic: {result.lr1}")
        print(f"Critical values (95%): {result.cvt[:, 1]}")
        for i in range(len(result.lr1)):
            if result.lr1[i] > result.cvt[i, 1]:
                print(f"  - r = {i}: Cointegration exists at 95% confidence")
            else:
                print(f"  - r = {i}: No cointegration at 95% confidence")
    except Exception as e:
        print(f"Could not perform cointegration test for {stock1} and {stock2}: {e}")

# Test cointegration for some pairs
cointegration_test_pair(data_df, 'AMD', 'INTC')
cointegration_test_pair(data_df, 'AMZN', 'TSLA')

# Forecast next 7 periods
forecast_steps = 7
forecasts = {stock: arima_models[stock].forecast(steps=forecast_steps) for stock in stocks}
forecast_index = range(len(data_df), len(data_df) + forecast_steps)

# Plot all graphs
# 1. Original series
plt.figure(figsize=(12, 6))
for stock in stocks:
    plt.plot(data[stock], label=stock)
plt.title('Portfolio Closing Prices (7 Days)')
plt.legend()
plt.show()

# 2. Differenced series
plt.figure(figsize=(12, 6))
for stock in stocks:
    plt.plot(diff_data[stock], label=f'{stock} Diff')
plt.title('Differenced Series (7 Days)')
plt.legend()
plt.show()

# 3. Combined forecast
plt.figure(figsize=(12, 6))
colors = {'AMD': 'red', 'INTC': 'green', 'AMZN': 'blue', 'NDAQ': 'orange', 'TSLA': 'purple'}
for stock in stocks:
    plt.plot(data[stock], label=f'{stock} Historical')
    plt.plot(forecast_index, forecasts[stock], label=f'{stock} Forecast', color=colors[stock])
plt.title('Portfolio Closing Prices with 7-Day Forecasts')
plt.legend()
plt.show()

# 4. Detailed forecast with confidence intervals
def plot_forecast(model, series, name, steps=7):
    forecast_obj = model.get_forecast(steps=steps)
    forecast = forecast_obj.predicted_mean
    conf_int = forecast_obj.conf_int()
    forecast_index = range(len(series), len(series) + steps)

    plt.figure(figsize=(12, 6))
    plt.plot(series, label=f'{name} Historical')
    plt.plot(forecast_index, forecast, label='Forecast', color='red')
    plt.fill_between(forecast_index, conf_int.iloc[:, 0], conf_int.iloc[:, 1],
                     color='pink', alpha=0.3, label='95% Confidence Interval')
    plt.title(f'{name} Price Forecast (7 Days)')
    plt.legend()
    plt.show()

    last_value = series.iloc[-1]
    mean_forecast = forecast.mean()
    print(f"\nForecast Interpretation for {name}:")
    print(f"Last observed value: {last_value:.2f}")
    print(f"Average forecast value: {mean_forecast:.2f}")
    print(f"Trend: {'Upward' if mean_forecast > last_value else 'Downward' if mean_forecast < last_value else 'Flat'}")

# Generate detailed forecast plots
for stock in stocks:
    plot_forecast(arima_models[stock], data[stock], stock)

# Print forecast values
for stock in stocks:
    print(f"\n{stock} Forecast Values (next 5 periods):")
    print(forecasts[stock][:5])
```
